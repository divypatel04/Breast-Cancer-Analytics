### Machine Learning
For this next section, we'll be implementing Random Forests to try and find gene interaction networks that can predict outcome. For this to work, we must aggregate based on genes. 

```python
def mutation_present(dataframe_genes, patient_id, gene):
    query_result = query_by_gene(dataframe_genes, gene)
    result = len(query_result[query_result['PATIENT_ID']== patient_id]) > 0
    return 0 if result == 0 else 1
```

```python
def create_gene_map(dataframe_genes, all_genes):
    gene_map = {}
    for gene in all_genes:
        patient_list = list(dataframe_genes[dataframe_genes['Hugo_Symbol']==gene]["PATIENT_ID"])
        gene_map[gene] = patient_list
    return gene_map
```

```python
gene_map = create_gene_map(clinical_data_with_patients, all_genes_mapped)
```

```python
def add_patient_gene_network(dataframe, mapped_genes, gene):
    temp_dataframe = dataframe.copy()
    patients = mapped_genes[gene]
    truth_value = lambda row: row["PATIENT_ID"] in patients
    label = lambda row:  1 if truth_value(row) else 0
    temp_dataframe[gene] = temp_dataframe.apply(label, axis="columns") #Apply Label
    return temp_dataframe
```

```python

for gene in all_genes_mapped:
    patients_gene_network = add_patient_gene_network(patients_gene_network, gene_map, gene)
patients_gene_network
```

```python
def create_table_of_analysis(dataframe, real_label_column_name, test_label_column_name, positive_result, negative_result):
    # Ensure proper data type comparisons
    positive_result = int(positive_result) if isinstance(positive_result, (str, bool)) and positive_result.isdigit() else positive_result
    negative_result = int(negative_result) if isinstance(negative_result, (str, bool)) and negative_result.isdigit() else negative_result

    # Calculate TP, FP, TN, FN using boolean masks
    tp_mask = (dataframe[real_label_column_name] == positive_result) & (dataframe[test_label_column_name] == positive_result)
    fp_mask = (dataframe[real_label_column_name] == negative_result) & (dataframe[test_label_column_name] == positive_result)
    tn_mask = (dataframe[real_label_column_name] == negative_result) & (dataframe[test_label_column_name] == negative_result)
    fn_mask = (dataframe[real_label_column_name] == positive_result) & (dataframe[test_label_column_name] == negative_result)

    # Count occurrences
    true_positive_r = tp_mask.sum()
    false_positive_r = fp_mask.sum()
    true_negative_r = tn_mask.sum()
    false_negative_r = fn_mask.sum()

    # Handle row count
    row_count = dataframe.shape[0]
    if row_count == 0:
        raise ValueError("Dataframe is empty; cannot compute metrics.")

    # Compute metrics
    accuracy = (true_positive_r + true_negative_r) / row_count
    tpr = true_positive_r / (true_positive_r + false_negative_r) if (true_positive_r + false_negative_r) > 0 else 0.0
    tnr = true_negative_r / (true_negative_r + false_positive_r) if (true_negative_r + false_positive_r) > 0 else 0.0

    # Create result dictionary
    d = {
        "TP": [true_positive_r],
        "FP": [false_positive_r],
        "TN": [true_negative_r],
        "FN": [false_negative_r],
        "Accuracy": [accuracy],
        "TPR": [tpr],
        "TNR": [tnr]
    }

    # Return as DataFrame
    result_table = pd.DataFrame(data=d)
    return result_table
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Function to plot ROC curve
def plot_roc_curves(y_test, rf_probs, svm_probs):
    # Calculate ROC curve and ROC area for Random Forest
    fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_probs)
    roc_auc_rf = auc(fpr_rf, tpr_rf)

    # Calculate ROC curve and ROC area for SVM
    fpr_svm, tpr_svm, _ = roc_curve(y_test, svm_probs)
    roc_auc_svm = auc(fpr_svm, tpr_svm)

    # Plot both ROC curves
    plt.figure(figsize=(10, 8))
    plt.plot(fpr_rf, tpr_rf, color='darkorange', lw=2,
             label=f'Random Forest ROC curve (AUC = {roc_auc_rf:.2f})')
    plt.plot(fpr_svm, tpr_svm, color='green', lw=2,
             label=f'SVM ROC curve (AUC = {roc_auc_svm:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curves')
    plt.legend(loc="lower right")
    plt.show()

    return roc_auc_rf, roc_auc_svm

# Prepare the data (assuming X_train, X_test, y_train, y_test are already defined)
# If not defined, uncomment and modify these lines:
# X = patients_gene_network.drop(['PATIENT_ID', 'Survived'], axis=1)
# y = patients_gene_network['Survived']
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Scale the features for SVM
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Random Forest
rf_model = RandomForestClassifier(max_depth=1, n_estimators=1, random_state=0)
rf_model.fit(X_train, y_train)

# Train SVM
svm_model = SVC(kernel='rbf', probability=True, random_state=0)
svm_model.fit(X_train_scaled, y_train)

# Get probability predictions
rf_probs = rf_model.predict_proba(X_test)[:, 1]
svm_probs = svm_model.predict_proba(X_test_scaled)[:, 1]

# Plot ROC curves and get AUC scores
roc_auc_rf, roc_auc_svm = plot_roc_curves(Y_test, rf_probs, svm_probs)

# Print classification metrics for both models
from sklearn.metrics import classification_report

print("\nRandom Forest Classification Report:")
rf_preds = rf_model.predict(X_test)
print(classification_report(Y_test, rf_preds))

print("\nSVM Classification Report:")
svm_preds = svm_model.predict(X_test_scaled)
print(classification_report(Y_test, svm_preds))

# Optional: Perform Grid Search for SVM hyperparameters
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto', 0.1, 1],
    'kernel': ['rbf', 'linear']
}

grid_search = GridSearchCV(SVC(probability=True), param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train_scaled, y_train)

print("\nBest SVM parameters:", grid_search.best_params_)
print("Best SVM score:", grid_search.best_score_)

# Train best SVM model
best_svm = grid_search.best_estimator_
best_svm_probs = best_svm.predict_proba(X_test_scaled)[:, 1]

# Plot updated ROC curves with best SVM model
plot_roc_curves(Y_test, rf_probs, best_svm_probs)
```

### Decision Tree Genes
We use a RegEx to help find all genes in the tree

```python
res = ""
import re
pattern = '([A-Z])\w+[0-9]*'
genes_of_interest = []
geneExpression = re.compile(pattern)

for tree_i in maximum_model.estimators_:
    text_representation = tree.export_text(tree_i, feature_names=curious_genes)
    
    lines = text_representation.split("\n")
    for line in lines:
        mo = geneExpression.search(line)
        if mo:
            genes_of_interest.append(mo.group())
    
genes_of_interest = list(set(genes_of_interest))
genes_of_interest
```

```python
from scipy.stats import fisher_exact
# Extract unique genes of interest from decision trees
genes_of_interest = []
for tree_i in maximum_model.estimators_:
    text_representation = tree.export_text(tree_i, feature_names=curious_genes)
    lines = text_representation.split("\n")
    for line in lines:
        mo = geneExpression.search(line)
        if mo:
            genes_of_interest.append(mo.group())

genes_of_interest = list(set(genes_of_interest))

# Perform pairwise Fisher's exact tests and generate heatmap
# Extract the mutation matrix for the top 20 genes
mutation_matrix = patients_gene_network[genes_of_interest]

# Perform pairwise Fisher's exact tests
gene_names = genes_of_interest
p_values = np.zeros((len(gene_names), len(gene_names)))

for i, gene_i in enumerate(gene_names):
    for j, gene_j in enumerate(gene_names):
        if i < j:  # Perform tests for upper triangle only
            # Create contingency table
            contingency_table = pd.crosstab(mutation_matrix[gene_i], mutation_matrix[gene_j])
            if contingency_table.shape == (2, 2):  # Ensure valid 2x2 table
                _, p = fisher_exact(contingency_table, alternative="two-sided")
            else:
                p = 1.0  # Assign high p-value for invalid cases
            p_values[i, j] = p
            p_values[j, i] = p  # Mirror value for lower triangle

# Transform p-values for heatmap (log10 scale)
p_values[p_values == 0] = 1e-10
log_p_values = -np.log10(p_values)
log_p_values[np.isinf(log_p_values)] = 0  # Replace -inf with 0 for plotting
np.fill_diagonal(log_p_values, 0)  # Set diagonal to 0

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    log_p_values,
    xticklabels=gene_names,
    yticklabels=gene_names,
    cmap="RdBu_r",  # Red-Blue colormap for mutual exclusivity/co-occurrence
    center=0,
    annot=False,  # Set True to annotate with values
    cbar_kws={"label": "log10 p-value"}
)

plt.title("Co-occurrence and Mutual Exclusivity Heatmap")
plt.xlabel("Genes")
plt.ylabel("Genes")
plt.show()
```

```python
from scipy.stats import chi2_contingency
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure survival status is mapped in the dataset
patients_gene_network['Survival_Status'] = patients_gene_network['Survived'].apply(lambda x: 'Deceased' if x == 0 else 'Living')

# Clinical factor to analyze
clinical_factor = 'Survival_Status'
subgroups = ['Deceased', 'Living']

# Prepare results matrix for chi-squared p-values
chi_squared_p_values = pd.Series(index=genes_of_interest, dtype=float)

for gene in genes_of_interest:
    # Create contingency table for chi-squared test
    contingency_table = pd.crosstab(patients_gene_network[gene], patients_gene_network[clinical_factor])
    if contingency_table.shape[0] > 1 and contingency_table.shape[1] > 1:
        _, p, _, _ = chi2_contingency(contingency_table)
    else:
        p = 1.0  # Assign high p-value if table is invalid
    chi_squared_p_values[gene] = p

# Convert p-values to log10 scale
log_p_values = -np.log10(chi_squared_p_values)
log_p_values.replace([np.inf, -np.inf], 0, inplace=True)  # Handle infinite values

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    log_p_values.to_frame(name=f'log10(p-value)'),
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",
    center=0,
    cbar_kws={"label": "-log10(p-value)"}
)

plt.title("Associations Between Gene Mutations and Survival Status")
plt.xlabel("Clinical Factor")
plt.ylabel("Genes")
plt.show()
```

```python
import scipy
assert(scipy.__version__ >= "1.10.1")


def converter(column):
    return [2 if item == 1 else 1 for item in column]

nominal_convert = train_naive_bayes[genes_of_interest].apply(lambda x: converter(x))
#print(train_naive_bayes[genes_of_interest])


from scipy.stats.contingency import association

association_data = []

for gene_i in genes_of_interest:
    data_row = []
    for gene_j in genes_of_interest:
        
        column_1 = nominal_convert[gene_i].to_numpy()
        column_2 = nominal_convert[gene_j].to_numpy()
        
        two_links = []
        for row_i in range(len(column_1)):
            two_links_temp = [column_1[row_i],column_2[row_i]]
            two_links.append(two_links_temp)
        
        data_row.append(association(two_links, method="tschuprow"))
    association_data.append(data_row)
    

df_association = pd.DataFrame(data = association_data, columns = genes_of_interest)
df_association
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, roc_auc_score, classification_report
from sklearn.model_selection import train_test_split

# Ensure curious_genes and genes_of_interest are defined
# Select only the features common to both train and test sets
common_features = genes_of_interest  # Ensure these are consistent

# Split the data using the common features
train, test = train_test_split(patients_gene_network, test_size=0.5, random_state=42, shuffle=True)

X_train = train[common_features].to_numpy()
Y_train = train['Survived'].to_numpy()
X_test = test[common_features].to_numpy()
Y_test = test['Survived'].to_numpy()

# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=0)
rf_model.fit(X_train, Y_train)
rf_probs = rf_model.predict_proba(X_test)[:, 1]

# Train SVM
svm_model = SVC(kernel='rbf', probability=True, random_state=0)
svm_model.fit(X_train, Y_train)
svm_probs = svm_model.predict_proba(X_test)[:, 1]

# Train Naive Bayes
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, Y_train)
nb_probs = naive_bayes.predict_proba(X_test)[:, 1]

# ROC and AUC Calculation
rf_auc = roc_auc_score(Y_test, rf_probs)
svm_auc = roc_auc_score(Y_test, svm_probs)
nb_auc = roc_auc_score(Y_test, nb_probs)

# Plot ROC Curves
rf_fpr, rf_tpr, _ = roc_curve(Y_test, rf_probs)
svm_fpr, svm_tpr, _ = roc_curve(Y_test, svm_probs)
nb_fpr, nb_tpr, _ = roc_curve(Y_test, nb_probs)

plt.figure(figsize=(10, 8))
plt.plot(rf_fpr, rf_tpr, label=f"Random Forest (AUC = {rf_auc:.3f})", color="darkorange", lw=2)
plt.plot(svm_fpr, svm_tpr, label=f"SVM (AUC = {svm_auc:.3f})", color="green", lw=2)
plt.plot(nb_fpr, nb_tpr, label=f"Naive Bayes (AUC = {nb_auc:.3f})", color="blue", lw=2)
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves for Random Forest, SVM, and Naive Bayes Models")
plt.legend(loc="lower right")
plt.show()

# Feature Importance Plot for Random Forest
feature_importances = rf_model.feature_importances_
top_features_idx = feature_importances.argsort()[-10:][::-1]  # Top 10 features
top_features = [common_features[i] for i in top_features_idx]
top_importances = feature_importances[top_features_idx]

plt.figure(figsize=(12, 6))
plt.barh(top_features, top_importances, color="skyblue")
plt.gca().invert_yaxis()
plt.title("Top 10 Gene Feature Importances in Random Forest Model")
plt.xlabel("Importance Score")
plt.ylabel("Gene")
plt.show()

# Print classification reports for all models
print("Random Forest Classification Report:")
print(classification_report(Y_test, rf_model.predict(X_test)))

print("\nSVM Classification Report:")
print(classification_report(Y_test, svm_model.predict(X_test)))

print("\nNaive Bayes Classification Report:")
print(classification_report(Y_test, naive_bayes.predict(X_test)))
```
